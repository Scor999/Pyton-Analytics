**2.1 Изучите распределение домов от наличия вида на набережную**       
*Постройте график*      
*Сделайте выводы*

```python
data3 = df['waterfront'].value_counts()
data3.head()
```

```python
plt.pie(data3, autopct='%1.1f%%', labels=data3.index, )
plt.title('распределение домов от наличия вида на набережную')
```

---

#### Вывод:  
Из гистограммы видно что большинство домов не имеет вид на набержную это $99,2\%$. И лишь всего $0,8\%$ домов имеют вид на набережную.

**2.2 Изучите распределение этажей домов.**

```python
data4 = df['floors'].value_counts()
data4.head()
```

```python
plt.figure(figsize=(4, 4))
sns.histplot(df['floors'], bins=15);
plt.title('распределение этажей домов.')
plt.xlabel('Кол-во этажей')
plt.ylabel('Кол-во домов')
```

---

#### Вывод:  
Из гистограммы видно что большинство домов одноэтажные. На втором месте по распределению находятся двух этажные дома.
